### 4.Image to Pencil Sketch with Python

```python
# Importing the required package
import cv2

import matplotlib.pyplot as plt
```

```python
#reading the image 
Img=cv2.imread("face.png")
```

```python
#Preview the image
cv2.imshow('Scenery',Img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 1. Convert from RGB to GreyScale
grey_image=cv2.cvtColor(Img, cv2.COLOR_BGR2GRAY)
```

```python
# 2. Inverting the Image
invert_image=cv2.bitwise_not(grey_image)
```

```python
# 3. blur the Image
blur_img=cv2.GaussianBlur(invert_image, (111,111),0)
```

```python
# 4.Inver the blur image
invblur_img=cv2.bitwise_not(blur_img)
```

```python
# 5.sketch the loaded image
sketch_img=cv2.divide(grey_image,invblur_img, scale=256.0)
```

```python
# Save it 
cv2.imwrite("sketch.png", sketch_img)
```

```python
# Dispaly the Sketch
cv2.imshow("Converted_img",sketch_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow("face", Img)
cv2.imshow("Converted_img", sketch_img)
cv2.waitKey(0)
```

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('RGB Color Schema',size=18)
plt.imshow(Img)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('BGR Color Schema',size=18)
RGB_img = cv2.cvtColor(Img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis('off')
```

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(RGB_img)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Sketch image', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
